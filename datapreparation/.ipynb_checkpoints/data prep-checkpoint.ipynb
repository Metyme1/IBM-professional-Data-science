# step5: Missing Value Treatment

```python
#choose the method of imputing missing value
#like mean ,median,mode 
```

```python
for i in ["BMI","Polio","Income composition of resources"]:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include="number").columns:
    df[i]=impute.fit_transform(df[[i]])
```

# Step:6 - outliers treatments

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3-1.5*iqr
    return lw,uw
```

```python
wisker(df['GDP'])
```

```python
for i in ['GDP','Total expenditure','Thinnes 1-19 years']:
    lw,uw=wisker(df[i])
```
